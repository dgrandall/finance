```python
try:
    l=float(input('What is the amount of the mortgage loan (answer in $USD)'))
except ValueError: 
    print('Please enter an number with or without a decimal')
```

```python
try:
    i= (float(input('What is the yearly rate of interest of the mortgage loan ( ex. for 5% answer 5)')))/1200
except ValueError: 
    print('Please enter an number with or without a decimal')
```

```python
try:
    n = (int(input('How long is the mortgage term in years')))*12
except ValueError: 
    print('Please enter an number with or without a decimal')
```

```python
r = (1/(1+i))
p = round((l*((1-r)/(r-r**(n+1)))),2)
 
print('this is the amount of money that was loaned to you:')
print("$ %f \n"%(round(l,2)))
 
print('this is the amount of your monthly payments:')
print("$ %s \n"%(round(p,2)))
 
total_paid = round((p * n),2)
print('the total you will pay is:')
print("$ %f \n"%(total_paid))
 
 
print('the total amount of interest you will pay is:')
total_interest = round((total_paid-l),2)
print("$ %f \n"%(total_interest))
```

```python
for month in range(n): 
    per_of_princ = round((((p * month)/l)*100),2)
    if per_of_princ <= 100:
        print("In month %d, you have paid %f percent of the principal."%(month+1,per_of_princ))
    else:
        print('OVERPAYING')
        print("In month %d, you have paid %f percent of the principal."%(month+1,per_of_princ))
```
